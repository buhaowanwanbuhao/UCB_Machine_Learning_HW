```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from keras.models import Sequential
from keras.utils import to_categorical
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from keras.callbacks import EarlyStopping
import joblib
```

# Read the CSV and Perform Basic Data Cleaning

```python
exoplanet_df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
exoplanet_df.dropna(axis='columns', how='all', inplace=True)
# Drop the null rows
exoplanet_df.dropna(inplace=True)
exoplanet_df.head()
```

```python
# Remove Space for `FALSE POSITIVE` category
mask = exoplanet_df["koi_disposition"] == "FALSE POSITIVE"
exoplanet_df.loc[mask, "koi_disposition"] = "False_Positive"
exoplanet_df["koi_disposition"]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python

output = exoplanet_df["koi_disposition"]
feature = exoplanet_df.drop(columns=["koi_disposition"])

X_train, X_test, y_train, y_test = train_test_split(feature, output,
                                                    random_state=1, stratify=output)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using LabelEncoder and MinMaxScaler

```python
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
print(y_train_categorical.shape)
```

# Train the Model

```python
dnn_model = Sequential()
dnn_model.add(Dense(units=100, activation='relu', input_dim=40))
dnn_model.add(Dense(units=100, activation='relu'))
dnn_model.add(Dense(units=3, activation='softmax'))
```

```python
# Compile and fit the model
dnn_model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
dnn_model.summary()
```

```python
# set early stopping as callback
callbacks = [EarlyStopping(monitor='val_loss', patience=2)]
dnn_model.fit(
    X_train_scaled,
    y_train_categorical,
    callbacks=callbacks,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
dnn_model_loss, dnn_model_accuracy = dnn_model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {dnn_model_loss}, Accuracy: {dnn_model_accuracy}")
```

```python
encoded_predictions = dnn_model.predict_classes(X_test_scaled[:5])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:5])}")
```

# Save the Model

```python
# save fitted model to file

filename = 'deep_learning.sav'
joblib.dump(dnn_model, filename)
```
